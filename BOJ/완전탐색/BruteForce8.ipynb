```python
# 싱기한 네자리 숫자
# 6679
# https://www.acmicpc.net/problem/6679

def convert(n, base) : 
    add = 0
    while n != 0 :
        add += n % base
        n //= base
    return add

for i in range(2992, 10000) : 
    n10 = convert(i, 10)
    n12 = convert(i, 12)
    n16 = convert(i, 16)

    if n10 == n12 == n16 : print(i)
```
